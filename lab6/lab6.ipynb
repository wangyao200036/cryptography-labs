```python
import math
import random
```

```python
def gcd(a,b):
    while b:
        a,b=b,a%b
    return a
```

```python
def pollards_rho(n,c=1,f=None,max_iterations=1000000):
    if n%2==0:
        return 2
    if f is None:
        f = lambda x:(x*x+1)%n

    a=c
    b=c

    for i in range(max_iterations):
        a=f(a)
        b=f(f(b))
        d=gcd(abs(a-b),n)

        if 1<d<n:
            return d
        if d==n:
            return None

    return None
```

```python
n=1359331
n2=10403
```

```python
factor=pollards_rho(n,c=1,f=lambda x:(x*x+5)%n)
factor2=pollards_rho(n2,c=1,f=lambda x:(x*x+5)%n2)
```

```python
if factor:
    print(f"{factor}")
    print(f"{n//factor}")
```

```python
if factor2:
    print(f"{factor2}")
    print(f"{n2//factor2}")
```

```python
def fermat_factorizaton(n,max_iterations=1000000):
    if n%2==0:
        return 2

    s=math.isqrt(n)
    if s*s==n:
        return s

    s+=1
    for i in range(max_iterations):
        t2=s*s-n
        t=math.isqrt(t2)

        if t*t==t2:
            p=s-t
            if p>1 and n%p==0:
                return p
            q=s+t
            if q<n and n%q==0:
                return q
            return p

        s+=1

        if s>(n+1)/2:
            break
    return None
```

```python
factor=fermat_factorization(n)
```

```python
factor2=fermat_factorization(n2)
```
